```python
from transformers import pipeline
import json
```

```python
generator = pipeline(
    "text-generation",
    model="distilgpt2"
)
```

```python
conversation = []
```

```python
def chat(user_message):

    conversation.append({
        "role": "user",
        "content": user_message
    })

    history = ""

    for message in conversation:
        history += f"{message['role']}: {message['content']}\n"

    prompt = history + "assistant:"

    response = generator(
        prompt,
        max_new_tokens=60,
        do_sample=True,
        temperature=0.7
    )

    answer = response[0]["generated_text"].split("assistant:")[-1].strip()

    conversation.append({
        "role": "assistant",
        "content": answer
    })

    return answer
```

```python
print(chat("My name is Geetha."))

print(chat("I am studying Artificial Intelligence."))

print(chat("What is my name?"))
```

```python
with open("conversation_history.json","w") as f:
    json.dump(conversation,f,indent=4)
```

```python
with open("conversation_history.json") as f:
    memory = json.load(f)

print(memory)
```

```python
while True:

    question = input("You: ")

    if question.lower() == "exit":
        break

    reply = chat(question)

    print("Bot:", reply)
```
