```python
from transformers import pipeline
```

```python
classifier = pipeline(
    "sentiment-analysis"
)
```

```python
text = "I absolutely love learning Large Language Models."

result = classifier(text)

print(result)
```

```python
reviews = [
    "This course is amazing!",
    "The movie was boring.",
    "The product quality exceeded my expectations.",
    "The service was terrible.",
    "Today is an ordinary day."
]

results = classifier(reviews)

for review, prediction in zip(reviews, results):
    print(f"Review : {review}")
    print(f"Sentiment : {prediction['label']}")
    print(f"Confidence : {prediction['score']:.4f}")
    print("-" * 40)
```

```python
while True:

    text = input("Enter a sentence (type 'exit' to quit): ")

    if text.lower() == "exit":
        break

    result = classifier(text)

    print("\nPrediction")
    print(result[0]["label"])
    print("Confidence:", round(result[0]["score"],4))
    print("-"*40)
```

```python
for sentence in reviews:

    prediction = classifier(sentence)[0]

    print(sentence)
    print(prediction["label"])
    print(prediction["score"])
    print()
```
