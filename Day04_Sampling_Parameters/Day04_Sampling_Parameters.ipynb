```python
from transformers import pipeline

generator = pipeline(
    "text-generation",
    model="distilgpt2"
)
```

```python
prompt = "The future of Artificial Intelligence is"
```

```python
output = generator(
    prompt,
    max_new_tokens=50,
    temperature=0.2,
    do_sample=True
)

print(output[0]["generated_text"])
```

```python
output = generator(
    prompt,
    max_new_tokens=50,
    temperature=0.7,
    do_sample=True
)

print(output[0]["generated_text"])
```

```python
output = generator(
    prompt,
    max_new_tokens=50,
    temperature=1.2,
    do_sample=True
)

print(output[0]["generated_text"])
```

```python
output = generator(
    prompt,
    max_new_tokens=50,
    do_sample=True,
    top_k=20
)

print(output[0]["generated_text"])
```

```python
output = generator(
    prompt,
    max_new_tokens=50,
    do_sample=True,
    top_p=0.9
)

print(output[0]["generated_text"])
```

```python
output = generator(
    prompt,
    max_new_tokens=50,
    temperature=0.8,
    top_k=50,
    top_p=0.95,
    do_sample=True
)

print(output[0]["generated_text"])
```
