```python
from transformers import pipeline
```

```python
generator = pipeline(
    "text-generation",
    model="distilgpt2"
)
```

```python
prompt = "Artificial Intelligence is transforming"

result = generator(
    prompt,
    max_new_tokens=50,
    temperature=0.7,
    do_sample=True
)

print(result[0]["generated_text"])
```

```python
prompts = [
    "The future of AI is",
    "Large Language Models can",
    "One day robots will",
    "Machine learning is"
]

for prompt in prompts:
    print(f"\nPrompt: {prompt}")
    output = generator(prompt, max_new_tokens=40, temperature=0.8)
    print(output[0]["generated_text"])
```
