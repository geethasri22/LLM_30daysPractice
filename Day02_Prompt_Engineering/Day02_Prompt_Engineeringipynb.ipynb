```python
from transformers import pipeline

generator = pipeline(
    "text-generation",
    model="distilgpt2"
)
```

```python
prompt = "Explain Artificial Intelligence."

result = generator(prompt, max_new_tokens=60)

print(result[0]["generated_text"])
```

```python
prompt = """
You are a university professor.

Explain Artificial Intelligence in simple terms.
"""

result = generator(prompt, max_new_tokens=100)

print(result[0]["generated_text"])
```

```python
prompt = """
Explain Artificial Intelligence to a 10-year-old child using an everyday example.
"""

result = generator(prompt, max_new_tokens=120)

print(result[0]["generated_text"])
```
